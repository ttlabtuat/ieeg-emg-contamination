```python
from boxplot_acc import concat_all_acc_tsv
from glob import glob
import pandas as pd

# ファイルパス取得＆DataFrame作成
overt_tsv_dir = 'results_sp_eog_2600'
overt_acc_paths = sorted(glob(f'{overt_tsv_dir}/spec70_140_eog/*_accuracy.tsv'))
overt_eog_acc = concat_all_acc_tsv(overt_acc_paths, task='overt')

covert_tsv_dir = 'results_sp_eog_2600_covert'
covert_acc_paths = sorted(glob(f'{covert_tsv_dir}/spec70_140_eog_covert/*_accuracy.tsv'))
covert_eog_acc = concat_all_acc_tsv(covert_acc_paths, task='covert')

# wide形式に変換：indexにsubject、列に['overt_HEOG', 'overt_VEOG', 'covert_HEOG', 'covert_VEOG']
overt_wide = overt_eog_acc.T.copy()
covert_wide = covert_eog_acc.T.copy()

anova_wide = pd.concat([overt_wide, covert_wide], axis=1)
anova_wide = anova_wide.reset_index().rename(columns={'index': 'subject'})

# csvとして出力
anova_wide.to_csv('eog_2wayanova_wide.csv', index=False)
anova_wide
```
