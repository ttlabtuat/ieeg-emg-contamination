```python
import pandas as pd
import numpy as np
import mne
import matplotlib
import matplotlib.pyplot as plt
import yaml

# --- ここで色リストを定義（region_listと同じ順番、または長さが同じになるように）---
# 例: lobe_list の順に色を指定（足りなければ追加、不要な分は削除すること！）
manual_region_colors = [
    "red",        # 0 Frontal
    "blue",       # 1 Parietal
    "green",      # 2 Temporal
    "purple",     # 3 Occipital
    "orange",     # 4 Limbic/Insula
    "yellow",     # 5 Cerebellum
    # "brown",      # 6
    "teal",       # 7
    "navy",       # 9
    "tomato",     # 10
    # "gold",       # 11
    "slateblue",  # 12
    "orchid",     # 13
    "magenta",    # 14
    "olive",      # 16
    "cyan",       # 15
    "crimson",    # 17
    "mediumseagreen", # 18
    "lightgreen",       # 19
]
# 上記を region_list の順と対応させる

# YAMLファイルからマッピングを読み込む
with open('region_map/aal_lobe_mapping_speech.yaml', 'r', encoding='utf-8') as f:
    mapping_data = yaml.safe_load(f)
aal_to_region = mapping_data['aal_to_lobe']
region_list = mapping_data['lobe_list']

# region_plot_order_new.yaml から region_anatomical_mapping を読み込む
with open('region_plot_order_anat.yaml', 'r', encoding='utf-8') as f:
    region_order_data = yaml.safe_load(f)
region_anatomical_mapping = region_order_data['region_anatomical_mapping']

# region_list, manual_region_colors の長さを合わせる
if len(manual_region_colors) < len(region_list):
    raise ValueError('manual_region_colors の色数が region_list より少ないです。色を追加してください')
elif len(manual_region_colors) > len(region_list):
    manual_region_colors = manual_region_colors[:len(region_list)]

region_to_color = dict(zip(region_list, manual_region_colors))

# 座標のCSVファイルを読み込む（label列が必要）
all_coord_csv = "all_coord_multi5_and_region.csv"
df = pd.read_csv(all_coord_csv)
ch_pos = df[['x', 'y', 'z']].values

# すでにregionがあれば削除
if 'region' in df.columns:
    df = df.drop(columns=['region'])

# regionを追加  
df['region'] = df['label'].map(aal_to_region)

# region列のカテゴリごとに色を割り当て（存在しないregionには黒を割り当て，Unknownはグレー）
def region_to_final_color(region):
    if pd.isna(region):
        return "#000000"
    elif region == "Unknown":
        return "gray"
    elif region == "Other":
        return "gray"
    return region_to_color.get(region, "#000000")

df['color'] = df['region'].map(region_to_final_color)

colors = df['color'].tolist()

# region対応の色凡例を region_list + df['region'].unique() のunion順で（region_list順優先）
unique_regions = df['region'].unique()
# region_list順（存在するもののみ）＋存在するがregion_listにないもの順
ordered_regions = [r for r in region_list if r in unique_regions]
remaining_regions = [r for r in unique_regions if r not in ordered_regions]
ordered_regions += list(remaining_regions)
region_color_map_for_df = {region: region_to_final_color(region) for region in ordered_regions}

# region名からabbr（略語）を取得する関数
def get_region_abbr(region):
    if region in region_anatomical_mapping:
        return region_anatomical_mapping[region].get('abbr', region)
    # スペースやかっこをアンダースコアに置換して検索
    region_key = region.replace(" ", "_").replace("(", "").replace(")", "")
    if region_key in region_anatomical_mapping:
        return region_anatomical_mapping[region_key].get('abbr', region)
    return region

# --- 論文用の凡例（Legend）を作成 ---
from matplotlib.patches import Patch

# フォント設定（論文向け）
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 10

# brain_regionsの順番でregionリストを作成
brain_regions_order = region_order_data['brain_regions']
ordered_regions_yaml = []
for lobe, regions in brain_regions_order.items():
    for region in regions:
        if region in unique_regions:
            ordered_regions_yaml.append(region)

# 凡例用のパッチとラベルを作成（brain_regionsの順番）
legend_elements = []
for region in ordered_regions_yaml:
    color = region_color_map_for_df.get(region, '#000000')
    abbr = get_region_abbr(region)
    legend_elements.append(Patch(facecolor=color, edgecolor='black', linewidth=0.5, label=abbr))

# 縦1列で凡例を作成
n_items = len(legend_elements)

# 凡例のみの図を作成
fig, ax = plt.subplots(figsize=(2.0, n_items * 0.35 + 0.4))
ax.axis('off')

# 凡例を配置（縦1列）
legend = ax.legend(
    handles=legend_elements,
    loc='center',
    ncol=1,
    frameon=True,
    framealpha=1.0,
    edgecolor='black',
    fancybox=False,
    fontsize=9,
    handlelength=1.2,
    handleheight=1.0,
    handletextpad=0.5,
    labelspacing=0.3,
)

# 凡例の枠線を細く
legend.get_frame().set_linewidth(0.8)

plt.tight_layout(pad=0.2)
# plt.savefig('region_legend.png', dpi=300, bbox_inches='tight', facecolor='white', transparent=False)
plt.savefig('修論修正/region_legend.svg', bbox_inches='tight', facecolor='white')
plt.show()
print("Saved: region_legend.png, region_legend.pdf")

# 結果をCSVに保存（オプション）
# df.to_csv(f"all_coord_and_lobe.csv", index=False)
```

```python
import numpy as np
import mne

import mne
from mne.datasets import fetch_fsaverage

# mat → csv
import scipy.io



# fsaverage データのパスを取得（なければダウンロードされる）
fs_dir = fetch_fsaverage(verbose=True)
subjects_dir = fs_dir.parent
subject = 'fsaverage'

# 電極名と座標（例：MNI 座標系，単位はメートル）
ch_names = [f"ch{i:02d}" for i in range(1, len(ch_pos) + 1)]
# MNI 座標を [m] 単位で指定
ch_pos_dict = dict(zip(ch_names, ch_pos))

# モンタージュの作成
# coord_frame='mri' は fsaverage における MNI 空間を指す
montage = mne.channels.make_dig_montage(ch_pos=ch_pos_dict, coord_frame='head')

# Raw / Epochs データに適用（ダミーデータ作成）
info = mne.create_info(ch_names=ch_names, sfreq=1000, ch_types='seeg')
raw = mne.io.RawArray(np.zeros((len(ch_pos), 1000)), info)
raw.set_montage(montage)

# 可視化
# trans='fsaverage' を指定することで座標系を自動調整
# 脳を上から見た図（'dorsal' view を利用）
brain = mne.viz.Brain(subject, subjects_dir=subjects_dir, surf='pial', alpha=0.5)
trans = mne.transforms.Transform('head', 'mri', np.eye(4))
brain.add_sensors(raw.info, trans=trans, sensor_colors=colors, sensor_scales=4)
views = ['dorsal', 'medial', 'ventral', 'lateral']
for view in views:
    brain.show_view(view)

    # 画像を保存（背景透過で保存）
    brain_image_path = f'fsaverage_{view}.png'
    brain.save_image(brain_image_path, mode='rgba')
    print(f'画像を保存しました（背景透過）: {brain_image_path}')
```
