```python
coh_csv = "coh_0_600_ep2600_right_overt_70_140_all_mean_usable_ch.csv"
acc_csv = "spec70_140_acc_bad_ch_removed.csv"

import pandas as pd

coh_df = pd.read_csv(coh_csv, index_col=0)
acc_df = pd.read_csv(acc_csv, index_col=0)

coh_melt = coh_df.reset_index().melt(id_vars='index', var_name='sub', value_name='coh')
coh_melt = coh_melt.rename(columns={'index': 'ch'})

acc_melt = acc_df.reset_index().melt(id_vars='index', var_name='sub', value_name='acc')
acc_melt = acc_melt.rename(columns={'index': 'ch'})

concat_df = pd.concat([coh_melt, acc_melt[['ch', 'sub', 'acc']]], axis=1)
# 重複する列（'ch'と'sub'）のうち、最初に現れたものを残して削除
concat_df = concat_df.loc[:,~concat_df.columns.duplicated()]
concat_df.dropna(subset=['coh', 'acc'], inplace=True)
concat_df
```
